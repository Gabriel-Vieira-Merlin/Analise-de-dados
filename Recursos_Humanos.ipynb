```python
#Libs Necessárias

#Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

#Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

#Lib para ignorar avisos
import warnings

#Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#Lendo a base de dados
base_dados = pd.read_csv('Salary_Data.csv')

#Verificando a leitura
base_dados.head()
```

```python
#Verificando a dimensão do dataset
base_dados.shape
```

```python
#Verificando os tipos de dados presentes
base_dados.info()
```

```python
#Verificando a existência de valores nulos
base_dados.isnull().sum()
```

```python
#Verificando a densidade de renda entre os funcionários
plt.title('Densidade do salário dos funcionários')
plt.xlabel('Salário')
plt.ylabel('Densidade')
sns.kdeplot(data=base_dados['Salary']);
```

```python
#Verificando a densidade de experiência entre os funcionários
plt.title('Densidade dos anos de experiência dos funcionários')
plt.xlabel('Anos de experiência')
plt.ylabel('Densidade')
sns.kdeplot(data=base_dados['YearsExperience']);
```

```python
#Plotando um gráfico scatter da relação entre salário e anos de experiência
sns.scatterplot(data=base_dados, x='Salary', y='YearsExperience')
```
