```python
#Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Lendo a base de dados
base_dados = pd.read_excel('Dados_Pib.xlsx')
```

```python
#Verificando os dados
base_dados.head(10)
```

```python
#Agrupando por estado e ano
base_dados.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
#Fazendo um sistema de grids com o Seaborn

#Criando o grid

#col define quantos grids terão com base no parâmetro passado, col_wrap ajusta os grids em quantas colunas forem passadas, hue define o que será plotado em cada gráfico
grid_grafico = sns.FacetGrid(base_dados, col='Territorialidades', col_wrap=4, hue='Territorialidades')

#Plotando um gráfico de linha que seja de Ano/PIB em cada gráfico do grid
grid_grafico = grid_grafico.map(plt.plot, 'Ano', 'PIB per capita', marker='o')

#Adicionando uma sombra abaixo da linha plotada
grid_grafico = grid_grafico.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2)

#Alterando os títulos para só mostra o nome da coluna
grid_grafico.set_titles('{col_name}')

#Definindo o título da grid
grid_grafico = grid_grafico.fig.suptitle('Evolução da Renda per capita por Estado \n Essse relatório foi elaborado no treinado "Python para Análise de Dados" \n Está dispónivel no canal do youtube @Data Viking', fontsize=18)
```
