```python
#Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Lendo a base de dados
base_dados = pd.read_excel('Dados_Empresas_Energia.xlsx')
```

```python
#Verificando as 5 primeiras entradas
base_dados.head()
```

```python
#Verificando a quantidade de dados presentes
base_dados.shape
```

```python
#Verificando o tipo de informação
base_dados.info()
```

```python
#Verificando a existência de valores nulos
base_dados.isnull().sum()
```

```python
#Definindo a data como índice
base_dados.set_index('Data', inplace=True)

#Verificando a mudança
base_dados
```

```python
#Grafico

#Estilo do gráfico
#plt.style.use('seaborn-darkgrid')

#Tamanho
plt.figure( figsize=(16, 7))

#Titulo
plt.title('Análise ações de empresas de Energia em relação a Petrobrás', loc='left', fontsize=18, fontweight=0)

#Plot da Petrobras
plt.plot(base_dados.index, base_dados['Petrobras'], color='#008c4a')
#Adicionando texto no fim da linha do gráfico
plt.text(base_dados.index[-1], base_dados['Petrobras'].tail(1), 'Petrobrás', color='#008c4a', horizontalalignment='left')

#Plot das outras empresas
for coluna in base_dados.columns[1:]:
  plt.plot(base_dados.index, base_dados[coluna], color='gray') #Plotando o gráfico
  plt.text(base_dados.index[-1], base_dados[coluna].tail(1), coluna, color='gray', horizontalalignment='left'); #Adicionando o texto no final das linhas
```
